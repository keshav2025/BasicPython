```python
# 6 Write a function to demonstrate the behavior of mutable and immutable arguments
def modify_arguments(num, num_list):
    num += 10
    print(f"Inside function - num: {num}")
    num_list.append(10)
    print(f"Inside function - num_list: {num_list}")
a = 5
b = [1, 2, 3]
print(f"Before function call - a: {a}, b: {b}")
modify_arguments(a, b)
print(f"After function call - a: {a}, b: {b}")
```

```python
 # 9 Write a Python program to convert user input from string to integer, float, and boolean types.
user_input = input("enter any value")
if user_input.isdigit():
    print("Integer:", int(user_input))
else:
    print("Not an integer")

if '.' in user_input and user_input.replace('.', '', 1).isdigit():
    print("Float:", float(user_input))
else:
    print("Not a float")

print("Boolean:", user_input.lower() == "true")
```

```python
# 10 Write code to demonstrate type casting with list elements
Strlist = ["10" "12" "14" "16"]
intList = [int(x) for x in Strlist]
print(intList)
FloatList = [float(x) for x in Strlist]
print(FloatList)
BoolList = [bool(x) for x in Strlist]
print(BoolList)
```

```python
# 14. Write a program to reverse a string using a while loop.
string = input("Enter any String : ")
reversed_string = ""
index = len(string) - 1
while index>=0:

  reversed_string += string[index]
  index -= 1
print(reversed_string)
```

```python
# 15 Write a Python program to calculate the factorial of a number provided by the user using a while loop.

num = int(input("Enter a any number  "))
f = 1
i = 1
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    while i <= num:
        f *= i
        i += 1
    print(f"The factorial of {num} is {f}")
```
